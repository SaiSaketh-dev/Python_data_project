```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.loc[0:10, 'salary_rate':'salary_hour_avg']
```

```python
salary_median_year_avg = df['salary_year_avg'].median()
salary_median_year_avg
```

```python
salary_median_hour_avg = df['salary_hour_avg'].median()
salary_median_hour_avg
```

```python
df_filled = df
df_filled['salary_year_avg'] = df_filled['salary_year_avg'].fillna(salary_median_year_avg)
df_filled['salary_hour_avg'] = df_filled['salary_hour_avg'].fillna(salary_median_hour_avg)
```

```python
df
```

```python
df_filled.loc[0:10, 'salary_year_avg':'salary_hour_avg']
```

```python
df_unique = df_filled
df_unique = df_unique.drop_duplicates()

print('original dataset:', len(df_filled))
print('revised dataset:', len(df_unique))
print('duplicates removed:', len(df_filled)-len(df_unique))
```

```python
df_unique = df_unique.drop_duplicates(subset=['job_title','company_name'])

print('original dataset:', len(df_filled))
print('revised dataset:', len(df_unique))
print('duplicates removed:', len(df_filled)-len(df_unique))
```
