```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_posted_month'] = df['job_posted_date'].dt.month
```

```python
job_usa = df[df['job_country'] == 'United States']
job_usa
```

```python
import calendar
job_usa_pivot = job_usa.pivot_table(index='job_posted_month', values= 'job_title', columns='job_title_short', aggfunc='count')
job_usa_pivot = job_usa_pivot.reset_index()
job_usa_pivot ['job_posted_month']= job_usa_pivot['job_posted_month'].apply(lambda x: calendar.month_name[x])
job_usa_pivot = job_usa_pivot.set_index('job_posted_month')
job_usa_pivot
```

```python
df_US_software_pivot = pd.read_csv('https://lukeb.co/software_csv').set_index('job_posted_month')
df_US_software_pivot
```

```python
us_merged = job_usa_pivot.merge(df_US_software_pivot, on='job_posted_month')
```

```python
top_5 = us_merged.sum().sort_values(ascending=False).head().index.to_list()
top_5
```

```python
us_merged[top_5]
```

```python
us_merged[top_5].plot(kind = 'line')
plt.title('US job trend for the data analysis and tech jobs in 2023')
plt.ylabel('Job Counts')
plt.xlabel('2023')
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
