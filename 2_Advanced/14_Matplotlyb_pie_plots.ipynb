```python
#importing libraries
import ast
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset

#loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

#cleanng data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind = 'pie', startangle = 90, autopct = '%1.1f%%')
plt.ylabel('')
plt.title('Work from home breakup')
plt.show()
```

```python
df_pie = df[['job_work_from_home', 'job_no_degree_mention', 'job_health_insurance']]
df_pie
```

```python
fig, ax = plt.subplots(1, 3)

for i, column in enumerate(df_pie.columns):
    df[column].value_counts().plot(kind = 'pie', startangle = 90, autopct = '%1.1f%%', ax=ax[i])
```

```python
job_titles = ['job_work_from_home', 'job_no_degree_mention', 'job_health_insurance']

fig, ax = plt.subplots(1, 3)

for i, job_title in enumerate(job_titles):
    df_pie[job_title].value_counts().plot(kind = 'pie', ax=ax[i], startangle = 90, autopct = '%1.1f%%', title= job_title)
    ax[i].set_ylabel('')

fig.tight_layout()
```
