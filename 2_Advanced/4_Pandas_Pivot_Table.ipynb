```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset


dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_shortened = df.pivot_table(index='job_country',values='salary_year_avg',columns='job_title_short', aggfunc='median')
df_shortened
```

```python
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_shortened[job_titles]
```

```python
top_six_countries = df['job_country'].value_counts().head(6).index
top_six_countries
```

```python
top_six_countries.index
```

```python
df_shortened = df.pivot_table(index='job_country',values='salary_year_avg',columns='job_title_short', aggfunc='median')
df_shortened = df_shortened.loc[top_six_countries]
job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']
df_shortened = df_shortened[job_titles]
df_shortened.plot(kind = 'bar')
plt.title('Top countries vs their median salaries')
plt.ylabel('Median Salary(USD)')
plt.xlabel('')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```
