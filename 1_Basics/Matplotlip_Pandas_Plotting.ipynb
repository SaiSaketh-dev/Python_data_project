```python
from datasets import load_dataset
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
dataset['train']
```

```python
df = dataset['train'].to_pandas()
```

```python
df
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
job_counts = df['job_title_short'].value_counts()
```

```python
job_counts.plot(kind = 'bar')
plt.title('Job counts by Job Title')
plt.ylabel('job counts')
plt.xlabel('')
plt.xticks(rotation = 45, ha = 'right')
plt.show()
```

```python
df.dropna(subset=['salary_year_avg'], inplace=True)
```

```python
df
```

```python
df.dropna(subset=['job_posted_date'], inplace=True)
```

```python
df
```

```python
df['job_posted_month'] = df['job_posted_date'].dt.month
df.plot(x= 'job_posted_month',y='salary_year_avg', kind='line')
```

```python
salary_job = df.groupby(['job_title_short'])['salary_year_avg'].median()
```

```python
salary_job
```

```python
salary_job = salary_job.sort_values()
salary_job
```

```python
salary_job.values
```

```python
plt.barh(salary_job.index, salary_job.values)
plt.title('job median salary')
plt.ylabel('')
plt.xlabel('median salary')
plt.show()
```

```python
salary_job.plot(kind = 'barh')
plt.title('job median salary')
plt.ylabel('')
plt.xlabel('median salary')
plt.show()
```
