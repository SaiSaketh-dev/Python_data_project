```python
from datasets import load_dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
```

```python
dataset
```

```python
df = dataset['train'].to_pandas()
```

```python
df
```

```python
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df.info()
```

```python
date_counts = df['job_posted_date'].value_counts()
```

```python
type(date_counts)
```

```python
date_counts
```

```python
date_counts.index
```

```python
date_counts.values
```

```python
date_counts = date_counts.sort_index()
date_counts
```

```python
plt.plot(date_counts.index,date_counts.values)
```

```python
df['job_posted_month'] = pd.to_datetime(df['job_posted_date']).dt.month
```

```python
df['job_posted_month']
```

```python
month_counts = df['job_posted_month'].value_counts()
month_counts
```

```python
month_counts = month_counts.sort_index()
month_counts
```

```python
plt.plot(month_counts.index, month_counts.values)
plt.show()
```

```python
job_count = df['job_title_short'].value_counts()
job_count
```

```python
job_count = job_count.sort_values(ascending=False)
job_count
```

```python
plt.bar(job_count.index, job_count.values)
plt.title('Job Posting by Count')
plt.ylabel('Job Count')
plt.xlabel('Jobs')
plt.xticks(rotation=45, ha='right')
plt.show()
```
